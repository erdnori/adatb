```sql
SELECT *,
       v.NEV AS 'Ügyfél neve',
       AVG(f.FELNOTT_SZAM+f.GYERMEK_SZAM) OVER(PARTITION BY v.USERNEV ORDER BY v.USERNEV
                                                ROWS BETWEEN 3 PRECEDING AND 1 PRECEDING) as 'Előző három foglalás átlaga'
       
FROM Vendeg v LEFT JOIN Foglalas f ON v.USERNEV=f.UGYFEL_FK
              LEFT JOIN Szoba sz ON f.SZOBA_FK=sz.SZOBA_ID
              LEFT JOIN Szallashely szh ON sz.SZALLAS_FK=szh.SZALLAS_ID
```

```sql
SELECT sz.SZOBA_ID as 'Szoba azonosító',
       f.FOGLALAS_PK as 'Foglalás azonosító',
       DATEDIFF(DAY, f.METTOL, f.MEDDIG) as 'Időtartam napokban',
       LAG(DATEDIFF(DAY, f.METTOL, f.MEDDIG), 1, 0) OVER(PARTITION BY sz.SZOBA_ID ORDER BY f.FOGLALAS_PK) AS 'Előző foglalásának időtartama'
FROM Szoba sz JOIN Foglalas f ON sz.SZOBA_ID=f.SZOBA_FK
```
