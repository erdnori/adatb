1\. Legkisebb, legnagyobb, illetve átlag listaár

```sql
SELECT MIN(LISTAAR) AS 'Minimum',
       MAX(LISTAAR) AS 'Maximum',
       AVG(LISTAAR) AS 'Átlag'
FROM Termek
WHERE LISTAAR IS NOT NULL
```

2\. Megmutatja hány olyan sor van a termékekre nézve, ahol van, illetve nincs leírás

```sql
SELECT CASE 
            WHEN LEIRAS IS NOT NULL THEN 'Van'
            ELSE 'Nincs'
        END AS 'Leírás',
        COUNT(*) AS 'DB'
FROM Termek
GROUP BY CASE 
            WHEN LEIRAS IS NOT NULL THEN 'Van'
            ELSE 'Nincs'
        END
```

3. Termékek számának eloszlása készlet szerint

```sql
SELECT CASE 
            WHEN KESZLET<=10000 THEN '0-10.000 DB'
            WHEN KESZLET<=20000 AND KESZLET>10000 THEN '10.000-20.000 DB'
            WHEN KESZLET<=30000 AND KESZLET>20000 THEN '20.000-30.000 DB'
            WHEN KESZLET<=40000 AND KESZLET>30000 THEN '30.000-40.000 DB'
            WHEN KESZLET>40000 THEN '40.000+ DB'
        END AS 'Készlet',
        COUNT(*) AS 'Termékek száma'
FROM Termek
WHERE KESZLET IS NOT NULL
GROUP BY CASE 
            WHEN KESZLET<=10000 THEN '0-10.000 DB'
            WHEN KESZLET<=20000 AND KESZLET>10000 THEN '10.000-20.000 DB'
            WHEN KESZLET<=30000 AND KESZLET>20000 THEN '20.000-30.000 DB'
            WHEN KESZLET<=40000 AND KESZLET>30000 THEN '30.000-40.000 DB'
            WHEN KESZLET>40000 THEN '40.000+ DB'
        END
    
```

4\. Top 10 leghosszabb nevű termék

```sql
SELECT TOP 10 MEGNEVEZES AS 'Név', 
        LEN(MEGNEVEZES) AS 'Hossz'
FROM Termek
GROUP BY MEGNEVEZES
ORDER BY 2 DESC
```
